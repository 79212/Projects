```
import pandas as pd
df = [0, 0, 0]
df = pd.DataFrame(df).transpose()
df.columns = ['A', 'B', 'C']
df
```

```
df['A'][0] = 100
df['B'][0] = 0
df['C'][0] = 0
df
```

```
for _ in range(1, 10):
  df.loc[_, 'A'] = df.loc[_-1, 'A']+1
  df.loc[_, 'B'] = df.loc[_, 'A']+100
  df.loc[_, 'C'] = df.loc[_-1, 'B']+500
df
```

```
df['D'] = 0
df
```

```
import statistics



for _ in range(1, 10):
  df.loc[_, 'A'] = df.loc[_-1, 'A']+1
  df.loc[_, 'B'] = df.loc[_, 'A']+100
  df.loc[_, 'C'] = df.loc[_-1, 'B']+500
  df.loc[_, 'D'] = statistics.mean([df.loc[_, 'B'], df.loc[_, 'C']])
df
```
