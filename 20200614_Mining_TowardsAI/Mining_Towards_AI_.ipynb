```
def req(URL):
  page = requests.get(URL)

  soup = BeautifulSoup(page.content, 'html.parser')

  job_elems = soup.find_all(class_='streamItem streamItem--postPreview js-streamItem')
  job_elems

  list1 = list()
  for link in job_elems:
    for m in link.find_all('a', href=True):
      list1.append(m['href'])
  list1

  list2 = list()
  for m in list1:
    if '@' not in m:
      if 'https://medium.com/towards-artificial-intelligence?source=' not in m:
        if '--responses' not in m:
          list2.append(m)
  list2 = list(dict.fromkeys(list2))
  return list2
```
