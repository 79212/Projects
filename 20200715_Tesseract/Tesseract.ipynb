```
import cv2
import pytesseract

pytesseract.pytesseract.tesseract_cmd = (
    r'/usr/bin/tesseract'
)

img_cv = cv2.imread(r'/content/one-does-not-simply-walk-into-mordor-except-this-guy-8594199 (1) (1).png')

# By default OpenCV stores images in BGR format and since pytesseract assumes RGB format,
# we need to convert from BGR to RGB format/mode:
d = cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)

print(pytesseract.image_to_string(d))
# OR
d = Image.frombytes('RGB', img_cv.shape[:2], img_cv, 'raw', 'BGR', 0, 0)
print(pytesseract.image_to_string(d))
```

```
import cv2
import pytesseract
from pytesseract import Output
img = cv2.imread(r'/content/one-does-not-simply-walk-into-mordor-except-this-guy-8594199 (1) (1).png')

d = pytesseract.image_to_data(img, output_type=Output.DICT)
print(d.keys())
```

```
from google.colab.patches import cv2_imshow
n_boxes = len(d['text'])
for i in range(n_boxes):
    if int(d['conf'][i]) > 60:
        (x, y, w, h) = (d['left'][i], d['top'][i], d['width'][i], d['height'][i])
        img = cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

cv2.imshow('img', img)
cv2.waitKey(0)
```
