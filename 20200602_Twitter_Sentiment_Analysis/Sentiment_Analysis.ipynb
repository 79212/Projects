```
#tokenizer
def format_sentence(sent):
  return({word: True for word in nltk.word_tokenize(sent)})
```

##Tweets

```
print(nltk.word_tokenize("The cat is very cute"))
```

```
#se chiamiamo a al di fuori di questo slot non funziona
total = open('/content/drive/My Drive/Colab Notebooks/Projects/20200602_Twitter_Sentiment_Analysis/pos_tweets.txt')
positive_tweets = list()
#word tokenization
for sentence in total:
  #print(sentence)
  positive_tweets.append([format_sentence(sentence), 'pos'])
  #saves the sentence in format: [{tokenized sentence}, 'pos]
positive_tweets
```

```
#se chiamiamo a al di fuori di questo slot non funziona
total = open('/content/drive/My Drive/Colab Notebooks/Projects/20200602_Twitter_Sentiment_Analysis/neg_tweets.txt')
negative_tweets = list()
#word tokenization
for sentence in total:
  #print(sentence)
  negative_tweets.append([format_sentence(sentence), 'neg'])
  #saves the sentence in format: [{tokenized sentence}, 'pos]
negative_tweets
```

```
train = positive_tweets[:int((.9)*len(positive_tweets))] + negative_tweets[:int((.9)*len(negative_tweets))]
test = positive_tweets[int((.1)*len(positive_tweets)):] + negative_tweets[int((.1)*len(negative_tweets)):]
```

```
print(train[0])
```

```
from nltk.classify import NaiveBayesClassifier

classifier = NaiveBayesClassifier.train(train)
classifier.show_most_informative_features()
```

```
example2 = "beautiful"
print(classifier.classify(format_sentence(example2)))
```

```
from nltk.classify.util import accuracy
print(accuracy(classifier, test))
```
